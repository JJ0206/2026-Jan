datetime.date 년, 월, 일로 날짜를 표현할 때 사용한다.


```python
import datetime

day1=datetime.date(2024,2,23)
print(day1)

day2=datetime.date(2025,12,31)
print(day2)
```

```python
import datetime

day1=datetime.date(2024,2,23)
print(day1)

day2=datetime.date(2025,12,31)
print(day2)

diff=day2-day1
print(diff)
```

```python
import datetime

day1=datetime.date(2024,2,23)
print(day1)

day2=datetime.date(2025,12,31)
print(day2)

diff=day2-day1
print(diff)

print(diff.days)
print(diff.seconds)
print(diff.total_seconds())
```

datetime.time은 년, 월, 일로만 구성되어 있다.

시, 분, 초까지 포함하려면 datetime.datetime을 사용해야 한다.

```python
import datetime
day3=datetime.datetime(2025,12,31,14,12,00)
print(day3)
print(day3.hour)
```

날짜 객체의 weekday() 메소드를 활용하면 요일을 숫자로 구할 수 있다.


0: 월요일
1: 화요일
2: 수요일
,,,
6: 일요일

```python
import datetime
day=datetime.date(2026,1,16)
print(day.weekday())
```

월요일을 1로 지정하고 싶다면 isoweekday() 메소드 사용


```python
import datetime
day=datetime.date(2026,1,16)
print(day.weekday())
print(day.isoweekday())
```

오늘 날짜를 구할 때는
datetime.date.today() 사용


```python
import datetime
today=datetime.date.today()
print(today)
```

n일 후의 날짜를 구하려면
datetime.timedelta(days=n) 사용

```python
import datetime
today=datetime.date.today()
diff=datetime.timedelta(days=100)
day2=today+diff
print(day2)
```

두 날짜 사이의 차이를 구하려면 뺄셈한다.

```python
import datetime
today=datetime.date.today()
dday=datetime.date(2026,3,14)
diff=dday-today
print(diff)
print(diff.days)
```
