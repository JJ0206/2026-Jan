datetime.date 년, 월, 일로 날짜를 표현할 때 사용한다.


```python
import datetime

day1=datetime.date(2024,2,23)
print(day1)

day2=datetime.date(2025,12,31)
print(day2)
```

```python
import datetime

day1=datetime.date(2024,2,23)
print(day1)

day2=datetime.date(2025,12,31)
print(day2)

diff=day2-day1
print(diff)
```

```python
import datetime

day1=datetime.date(2024,2,23)
print(day1)

day2=datetime.date(2025,12,31)
print(day2)

diff=day2-day1
print(diff)

print(diff.days)
print(diff.seconds)
print(diff.total_seconds())
```

날짜 객체의 weekday() 메소드를 활용하면 요일을 숫자로 구할 수 있다.


0: 월요일
1: 화요일
2: 수요일
,,,
6: 일요일

```python
import datetime
day=datetime.date(2026,1,16)
print(day.weekday())
```

월요일을 1로 지정하고 싶다면 isoweekday() 메소드 사용


```python
import datetime
day=datetime.date(2026,1,16)
print(day.weekday())
print(day.isoweekday())
```

오늘 날짜를 구할 때는
datetime.date.today() 사용


```python
import datetime
today=datetime.date.today()
print(today)
```

n일 후의 날짜를 구하려면
datetime.timedelta(days=n) 사용

```python
import datetime
today=datetime.date.today()
diff=datetime.timedelta(days=100)
day2=today+diff
print(day2)
```

두 날짜 사이의 차이를 구하려면 뺄셈한다.

```python
import datetime
today=datetime.date.today()
dday=datetime.date(2026,3,14)
diff=dday-today
print(diff)
print(diff.days)
```

**날짜 포맷**

%Y,4자리 연도 (Year),2026

%y,뒤의 2자리 연도,26

%m,0을 채운 월 (Month),01

%B,월의 전체 이름 (Locale 기준),January

%b,월의 약어 (Locale 기준),Jan

%d,0을 채운 일 (Day),16

```python
from datetime import datetime

now = datetime.now()

# 예: 2026년 01월 16일 (금) 13시 04분
formatted_date = now.strftime("%Y년 %m월 %d일 (%a) %H시 %M분")
print(formatted_date)

# ISO 8601 표준 형식과 유사하게 출력
iso_style = now.strftime("%Y-%m-%d %H:%M:%S")
print(iso_style)
```

**시간 포맷**

%H,24시간 형식 시간 (00~23),13

%I,12시간 형식 시간 (01~12),01

%p,오전/오후 (AM 또는 PM),PM

%M,0을 채운 분 (Minute),04

%S,0을 채운 초 (Second),30

%f,마이크로초 (6자리),000123

**요일 및 기타 정보**

%A,요일 전체 이름,Friday

%a,요일 약어,Fri

%w,요일을 숫자로 표시 (0: 일요일 ~ 6: 토요일),5

%j,연중 일수 (001~366),016

%Z,시간대 이름 (Timezone),KST

%%,% 문자 자체를 표시,%

```python
from datetime import datetime

# 기준 시간 설정 (2026년 1월 16일 13시 8분 59초)
now = datetime(2026, 1, 133, 13, 8, 59)

formats = [
    "%Y-%m-%d", "%y/%m/%d", "%H:%M:%S", "%I:%M:%S %p",
    "%A, %B %d", "%a, %b %d", "%j번째 날", "%W주차", "%c"
]

for fmt in formats:
    print(f"{fmt:<15} => {now.strftime(fmt)}")
```
